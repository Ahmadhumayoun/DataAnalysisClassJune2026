```python
choice = input("login or signup 1/2 : ")
if choice == 1:
    mail = input("Enter mail : ")
    passwd = input("entr passwd : ")
    with open("UsersDatabase.txt","r") as userFetchedDB:
        if mail in dict(userFetchedDB)["Names"] and passwd in dict(userFetchedDB)["Passwd"]:
            print("user logged in ")
        else:
            print("User doesn't exist")
elif choice == 2 :
    mail = input("Enter mail : ")
    passwd = input("entr passwd : ")
    with open("UsersDatabase.txt","r") as userFetchedDB:
        if mail in dict(userFetchedDB)["Names"] and passwd in dict(userFetchedDB)["Passwd"]:
            print("user Already exist ")
        else:
            with open("UsersDatabase.txt","w")  as newFile2Write:
                dictionaryData = dict(userFetchedDB)
                dictionaryData["Names"].append(mail)
                dictionaryData["Passwd"].append(passwd)
                data2Write = str(dictionaryData)
                newFile2Write.write(data2Write)


```
